* Adding

```python
def add(a, b):
    return a + b

print (add(2, 3))
```

* Even or Odd

```python
def even_or_odd(n):
    if n % 2 == 0:
        return "even"
    else:
        return "odd"

print(even_or_odd(5))
print(even_or_odd(10))
```

* Largest of Three

```python
def largest(a, b, c):
    if (a >= b) and (a >= c):
        return a
    elif (b >= a) and (b >= c):
        return b
    else:
        return c

print(largest(18, 9, 1991))
```

* Count Characters

```python
def count_chars(word):
    count = 0
    
    for i in word:
        count = count + 1
        #count = count += 1
    return count

print(count_chars("bird"))
print(count_chars("macbook"))
```

* First and Last Letter

```python
def first_last(word):
    first = word[0]
    last = word[-1]
    fnl = first + last
    return fnl
    #return word[0] + word[-1]

print(first_last("python"))
print(first_last("butterfly"))
```

### Part II -- Lists and Loops

* Sum of List

```python
def sum_list(random_list):
    count = 0
    
    for i in random_list:
        count = count + i
    return count

print(sum_list([1,2,3]))
print(sum_list([13,7,79]))
```

* Count Positives

```python
def count_positives(numbers):
    count = 0
    
    for i in numbers:
        if i >= 0:
            count += 1
    return count

print(count_positives([-1,2,4,-3]))
print(count_positives([8,-6,-9,-11]))
```

* Reverse List

```python
def reversed_list(numbers):
    new_list = []
    
    for i in numbers[::-1]:
        new_list.append(i)
    return new_list
    #return numbers[::-1]

print(reversed_list([1,2,3]))
print(reversed_list([90,50,70]))
```

* Find Maximum

```python
def max_num(numbers):
    largest = numbers[0]
    
    for i in numbers:
        if i > largest:
            largest = i
    return largest

print(max_num([5,3,9,1]))
```

* Remove Negatives

```python
def remove_negatives(numbers):
    for i in numbers:
        if i < 0:
            numbers.remove(i)
    return numbers

print(remove_negatives([-1,2,-3,4]))
```

### Part III -- Strings

* Count Vowels

```python
def count_vowels(word):
    count = 0
    vowels = "aAeEiIoOuU"
    
    for i in word:
        if i in vowels:
            count += 1
    return count
    
print(count_vowels("hello"))
```

* Reverse String

```python
def reversed_string(word):
    reversed_word = ""

    for i in word:
        reversed_word = i + reversed_word
    return reversed_word

print(reversed_string("cat"))
print(reversed_string("butterfly"))
```

* Palindrome Check
  ###### Return True if word reads same backwards

```python
def is_palindrome(word):
    reversed_word = ""
    
    for i in word:
        reversed_word = i + reversed_word
    return reversed_word == word

print(is_palindrome("racecar"))
print(is_palindrome("telephone"))
print(is_palindrome("ana"))
```

* Remove Spaces

```python
def remove_spaces(text):
    no_spaces = ""
    for i in text:
        if i != " ":
            no_spaces += i
    return no_spaces

print(remove_spaces("hi there"))
```

* Capitalize Words

```python
def capitalize_words(text):
    result = ""
    words = text.split(" ")
    
    for i in words:
        result += i[0].upper() + i[1:] + " "
    return result.strip()

print(capitalize_words("hello world"))
print(capitalize_words("once upon a time"))
```

### Part IV

* FizzBuzz

```python
def fizzbuzz(a, b):
    for number in range(a, b +1):
        if number % 3 == 0:
            print("Fizz")
        elif number % 5 == 0:
                print("Buzz")
        elif number % 3 == 0 and number % 5 == 0:
            print("FizzBuzz")
        else:
            print(number)

fizzbuzz(1,10)
```

* Count Ocurrences

```python
def count_ocurrences(numbers_list, num):
    count = 0
    
    for i in numbers_list:
        if i == num:
            count += 1
    return count

print(count_ocurrences([1,2,2,3,2],2))
```

* Second Largest

```python
def second_largest(num_list):
    largest = num_list[0]
    second = num_list[0]

    for num in num_list:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num
    return second

print(second_largest([5,1,9,7]))
```

* Unique Values

```python
def unique(nums):
    new_list = []
    
    for i in nums:
        if i in new_list:
            continue
        new_list.append(i)
    return new_list

print(unique([1,2,2,3]))
```

* Word Frequency

```python
def word_freq(words):
    new_dict = {}
    words_list = words.split(" ")
    
    for i in words_list:
        if i not in new_dict:
            new_dict[i] = 1
        else:
            new_dict[i] += 1
    return new_dict

print(word_freq("hi hi bye"))
```

### Part V-- Lists + Loops + Logic

* Sum of Even Numbers

```python
def sum_even(numbers):
    count = 0
    
    for i in numbers:
        if i % 2 != 0:
            continue
        else:
            count += i
    return count
    # sum(i for i in numbers if i % 2 == 0)

print(sum_even([1,2,3,4,5,6]))
```

* Smaller Number

```python
def smallest(numbers):
    smallest_num = numbers[0]
    
    for i in numbers:
        if i < smallest_num:
            smallest_num = i
    return smallest_num
    
print(smallest([5,2,9,1]))
print(smallest([13,7,99,101]))
```

* Count Letters

```python
def count_letters(word):
    new_dict = {}
    
    for i in word:
        if i not in new_dict:
            new_dict[i] = 1
        else:
            new_dict[i] += 1
    return new_dict

print(count_letters("hello"))
print(count_letters("racecar"))
```

* Numbers greater than N

```python
def greater_than(numbers, n):
    greater_nums = []
    
    for i in numbers:
        if i > n:
            greater_nums.append(i)
    return greater_nums

print(greater_than([1,5,8,2], 4))
```

* Multiply List

```python
def multiply(numbers):
    result = 1
    
    for i in numbers:
        result = result * i
    
    return result
print(multiply([1,2,3,4]))
```

* Find Index

```python
def find_index(numbers_list, target):
    for i in range(len(numbers_list)):
        if numbers_list[i] == target:
            return i
    #for index, value in enumerate(numbers_list):
    # if value == target:
    # return index

print(find_index([4,7,9], 7))
```

* Remove Duplicates, Preserve Order

```python
def remove_dupes(num_list):
    new_list = []
    
    for i in num_list:
        if i not in new_list:
            new_list.append(i)
    return new_list
print(remove_dupes([1,2,2,3,1]))
```

* All positives

```python
def all_positives(numbers):
    for i in numbers:
        if i <= 0:
            return False
    return True
    #return all(i > 0 for i in numbers)
    
print(all_positives([-1, 1, 1]))
print(all_positives([4,5,6]))
print(all_positives([0,1,2]))
```
